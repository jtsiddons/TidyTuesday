```python
def group_summary(df: pd.DataFrame, group_column: list[str], summary_column: str, num_values: int)-> pd.DataFrame:
    return df.query(f'{summary_column} > 0') \
        .groupby(group_column) \
        .agg(
            mean_val = pd.NamedAgg(column=summary_column, aggfunc="mean"),
            min_val = pd.NamedAgg(column=summary_column, aggfunc="min"), 
            quant_25 = pd.NamedAgg(column=summary_column, aggfunc=lambda x: np.quantile(a=x, q=0.25)),
            median = pd.NamedAgg(column=summary_column, aggfunc=np.median), 
            quant_75 = pd.NamedAgg(column=summary_column, aggfunc=lambda x: np.quantile(a=x, q=0.75)), 
            max_val = pd.NamedAgg(column=summary_column, aggfunc="max"), 
            count = pd.NamedAgg(column=summary_column, aggfunc="count"),
            ) \
        .sort_values('count', ascending=False) \
        .head(num_values)
```

```python
x = group_summary(
    horror_movies_df.query('vote_count >= 10'), 
    ['collection_name', 'original_language'], 
    'vote_average', 
    15
) \
    .sort_values('count', ascending=False)
```
